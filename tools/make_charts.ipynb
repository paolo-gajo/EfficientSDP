```python
import os
import json
walk_path = '/home/pgajo/projects/def-hsajjad/pgajo/Multitask-RFG-torch/results_steps'
entries = []
lens = []
for root, dirs, files in os.walk(walk_path):
    for D in dirs:
        dirname = os.path.join(root, D)
        file_list = [f for f in os.listdir(dirname) if os.path.isfile(os.path.join(dirname, f))]
        if file_list:
            
            load_path_results = os.path.join(dirname, 'val_results.json')
            load_path_bench = os.path.join(dirname, 'test_results_benchmark.json')
            
            with open(load_path_results, 'r', encoding='utf8') as f:
                data = json.load(f)
            with open(load_path_bench, 'r', encoding='utf8') as f:
                bench = json.load(f)
            model_tags = os.path.join(root, D).split('/')
            model_name = '_'.join([
                # model_tags[0],
                model_tags[8],
                model_tags[14],
                model_tags[18],
                model_tags[19],
            ])
            entry = {
                'model_name': model_name,
                'val_results': [el['parser_labeled_results']['F1'] for el in data],
                'learnable params': bench['learnable params']
            }
            lens.append(len(entry['val_results']))
            entries.append(entry)
max_len = max(lens)

# for i in range(len(entries)):
#     entries[i]['val_results'] += [0 for _ in range(max_len - len(entries[i]['val_results']))]
import matplotlib.pyplot as plt

# Plot setup
plt.figure(figsize=(12, 6))

# Iterate over entries and plot each line
for entry in entries:
    if 'arc_norm_0' in entry['model_name'] and 'freeze_encoder_0' in entry['model_name']:
        plt.plot(entry['val_results'], label=(entry['model_name'], f"{round(entry['learnable params']/1e6, 2)}M"))

# Axes labels and title
plt.xlabel('Step')
plt.ylabel('F1 Score')
plt.title('Validation F1 Scores per Step')
plt.legend(loc='lower right', fontsize='small')
plt.grid(True)
plt.tight_layout()

plt.show()
```
